```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol for S&P 500
ticker_symbol = 'SPY'

# Define the time period for the data
period = '7d'

# Define the interval for the data
interval = '1m'

# Download the data from Yahoo Finance
data = yf.download(ticker_symbol, period=period, interval=interval)

# Convert the index to a datetime object without timezone) in Eastern Standard Time
# 1. Convert to Eastern Time (ET). yfinance generates data in UTC time zone
data.index = data.index.tz_convert("US/Eastern")
# 2. Make the DatetimeIndex timezone-unaware for excel sheet***
data.index = data.index.tz_localize(None) # Removes the timezone from the index
df = pd.DataFrame(data)

#Print the first few and last few entries
print(data.head())  # Print the first few rows
print(data.tail())  # Print the last few rows

# Get the start and end dates from the dataframe
start_date = df.index.min().strftime('%Y-%m-%d')
end_date = df.index.max().strftime('%Y-%m-%d')

# Create an Excel sheet from the data
df.to_excel(f'data/generated_data/{ticker_symbol}_yfinance_7d_{interval}_{start_date}_to_{end_date}.xlsx', index=True)
```
